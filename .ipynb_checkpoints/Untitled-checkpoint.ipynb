```python
import IPython
%matplotlib notebook
from scipy.io import wavfile
samplerate, data = wavfile.read("c.wav")
print(f"number of channels = {data.shape[1]}")

length = data.shape[0] / samplerate
print(f"length = {length}s")

import matplotlib.pyplot as plt
import numpy as np
time = np.linspace(0., length, data.shape[0])
plt.plot(time, data[:, 0], label="Left channel")
plt.plot(time, data[:, 1], label="Right channel")
plt.legend()
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.show()
```

```python
d = data[:, 0][23000:]
plt.figure(figsize=(10,10))
plt.plot(d, label="Left channel")
plt.tight_layout()
plt.grid()
```

```python
freq = 261.6256 
samplerate * (1 / freq)
```

```python
import numpy as np
from scipy import signal
NFFT = 1024 * 2
s = d[20000:20000+NFFT]
windowed = s * signal.get_window("hamming", NFFT)
power = np.abs(np.fft.fft(windowed, NFFT) / NFFT) 
ff = np.fft.fftshift(power)
plt.figure()
plt.plot(10 * np.log10(ff))
plt.grid()
```

```python
from scipy import signal
f, Pxx_spec = signal.periodogram(d, samplerate, 'flattop', scaling='spectrum')
plt.figure()
plt.semilogy(f, np.sqrt(Pxx_spec))
# plt.ylim([1e-4, 1e1])
plt.xlabel('frequency [Hz]')
plt.ylabel('Linear spectrum [V RMS]')
plt.grid()
plt.show()
```

```python
f, Pxx_spec = signal.welch(d, samplerate, 'flattop', 1024*2, scaling='spectrum')
plt.figure()
plt.semilogy(f, np.sqrt(Pxx_spec))
plt.xlabel('frequency [Hz]')
plt.ylabel('Linear spectrum [V RMS]')
plt.show()
```

```python
def plot_response(fs, w, h, title):
    "Utility function to plot response functions"
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(0.5*fs*w/np.pi, 20*np.log10(np.abs(h)))
    ax.set_ylim(-80, 5)
    ax.set_xlim(0, 0.5*fs)
    ax.grid(True)
    ax.set_xlabel('Frequency (Hz)')
    ax.set_ylabel('Gain (dB)')
    ax.set_title(title)
    
fs = samplerate        # Sample rate, Hz
cutoff = 10    # Desired cutoff frequency, Hz
trans_width = 10  # Width of transition from pass band to stop band, Hz
numtaps = 1024 * 2 * 2 * 2 * 2 * 2     # Size of the FIR filter.
taps = signal.remez(numtaps, [0, cutoff, cutoff + trans_width, 0.5*fs], [1, 0], Hz=fs)
w, h = signal.freqz(taps, [1], worN=20000)
plot_response(fs, w, h, "Band-pass Filter")
```

```python
f, t, Sxx = signal.spectrogram(d, samplerate, window="hamming", nperseg=1024, noverlap=512)
print(Sxx.shape)
plt.figure()
plt.pcolormesh(t, f, Sxx, shading='gouraud')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```python
import librosa
from librosa import display
import numpy as np
y = np.array(d)
D = librosa.stft(np.array(d))  # STFT of y
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)

plt.figure()
librosa.display.specshow(S_db)
plt.colorbar()
```

```python
fig, ax = plt.subplots()
y = np.array(d)
D_highres = librosa.stft(y, hop_length=128, n_fft=1024)
S_db_hr = librosa.amplitude_to_db(np.abs(D_highres), ref=np.max)
img = librosa.display.specshow(S_db_hr, hop_length=128, x_axis='time', y_axis='log',
                               ax=ax)
ax.set(title='Higher time and frequency resolution')
# fig.colorbar(img, ax=ax, format="%+2.f dB")
```
